### 1. Import the necessary libraries if you are starting a new notebook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load the csv. Use the variable customer_df as customer_df = pd.read_csv().

```python
customer_df = pd.read_csv("files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
customer_df
```

### 3. What should we do with the customer_id column?

```python
### Drop customer column
customer_df = customer_df.drop(['Customer'], axis = 1)
```

```python
customer_df
```

### 4. Load the continuous and discrete variables into numericals_df and categorical_df variables, for eg.

```python
numerical_df = customer_df.select_dtypes(include=np.number)
categorical_df = customer_df.select_dtypes(include=np.object)
```

```python
numerical_df.columns
```

```python
categorical_df.columns
```

### 5. Plot every categorical variable

```python
for col in categorical_df.columns:
    categorical_df[col].value_counts().plot(kind='bar', xlabel=[col], ylabel=['Customer Lifetime Value'], rot=0)
    plt.show()
```

```python
# Gender has similar distribution for CLV
# Medsize and Four Car vehicles have higher CLV
# Web sales channel has the smaller CLV results
```
